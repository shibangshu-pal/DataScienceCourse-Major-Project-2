```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/akpythonyt/Datasets/main/matches.csv')
#sourced the data from - https://github.com/akpythonyt/Datasets/blob/main/matches.csv
df
```

```python
df.to_csv('Matches Dataset.csv')
```

```python
import numpy as np
team_uni = np.array(df['team1'].unique())
print(team_uni)
```

```python
df.info()
```

```python

df = df.drop(columns = ['season','city','date','toss_decision','result','dl_applied','win_by_runs','win_by_wickets','player_of_match','venue','umpire1','umpire2','umpire3'])
#deleted unncessary columns
```

```python
#eliminating empty rows for the data
df =df.dropna()
df
```

```python
df.to_csv('Modified_data.csv')
```

```python
t1 = np.array(df['team1'].values)
t2 = np.array(df['team2'].values)
t_win = np.array(df['toss_winner'].values)
win = np.array(df['winner'].values)

#team_uni contains the names of the teams uniquely
for i in range(752):
  for j in range(14):
    if t1[i] == team_uni[j]:
      t1[i] = j
    if t2[i] == team_uni[j]:
      t2[i] = j
    if t_win[i] == team_uni[j]:
      t_win[i] = j
    if win[i] == team_uni[j]:
      win[i] = j
print(t1)
print(t2)
print(t_win)
print(win)  
```

```python
arr_res = []
for i in range(752):
  if t_win[i] == win[i]:
    arr_res.append([t1[i],t2[i],t_win[i],win[i],1])
  else:
    arr_res.append([t1[i],t2[i],t_win[i],win[i],0])
arr_res
```

```python
df2 = pd.DataFrame(columns = ['Team1','Team2','Toss_winner','Match_winner','Pred'],data =arr_res)
df2
```

```python

x=df2.iloc[:,2:4].values   #input
```

```python
#initial visualisation, before clustering
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (10,4))
sns.scatterplot(df2['Toss_winner'],df2['Match_winner'])
```

```python

#Here I fix the number of clusters to be 4 
#now I use clusterer
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2, random_state = 0)  # there are two clusters, toss won and match won, toss won and match lost
model.fit(x)
y=model.predict(x)   #output
```

```python
#visualisation
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (10,4))
for i in range(1,2):
  sns.scatterplot(x[y==i,0],x[y == i,1],color='orange')
  
```

```python
#visualisation
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (10,4))
for i in range(0,2,2):
  sns.scatterplot(x[y==i,0],x[y == i,1])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (10,4))
for i in range(2):
  plt.scatter(x[y==i,0],x[y == i,1],label = f'Cluster {i}')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s= 300,c = 'yellow',label ='centroids')
```
